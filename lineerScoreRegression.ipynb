4 ) Veriyi uygun hale getirmek

Verilerimdeki değişkenleri x ve y değişkenlerine tanımladık.
bu tanımlama ile sklearn kütüphanesini birleştiricez

```python
x = df["Hours"]
print(x)
```

```python
y = df["Scores"]
print(y)
```

sklearn.model_selection.train_test_split ile verilerimizi yollayıp hangi oranda eğiticeğimizi söyledik(eğer verilmezse .25 .75 oranında yapar)


```python
import sklearn.model_selection 
```

```python
x_train , x_test, y_train, y_test = sklearn.model_selection.train_test_split(x,y,train_size = 0.8 )
```

```python
print(x_train)
```

```python
print(y_train)
```

Şuan x_train, x_test, y_test ve y_train (5,) tipinde bir array
bunu eğitmek için (5,1) yapmamız gerekiyor
bunuda numpy kütüphanesini kullanarak
1)array'i numpyArray olarak yap
// reshape(satirsayisi,sütun sayısı)

2)numpyArray olduktan sonra reshape(-1,1) kullanarak (x,1) lik oluştur.


```python
x_test.shape
```

```python
import numpy as np
```

```python
x_test = np.array(x_test)
x_train = np.array(x_train)
y_test = np.array(y_test)
y_train = np.array(y_train)
```

```python
x_test.shape
```

```python
x_test
```

```python
type(x_test)
```

```python
x_test = x_test.reshape(-1,1)
x_train = x_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
y_train = y_train.reshape(-1,1)
```

```python
type(x_test)
```

```python
x_test.shape
```

```python
print(x_train)
x_train.shape
```

```python
x_test
```

5) Model seçimi ve modelin eğitilmesi

```python
import sklearn.linear_model
```

```python
lin_model = sklearn.linear_model.LinearRegression()
```

```python
lin_model.fit(x_train,y_train)
```

6)Modelin optimizasyonu

```python
import sklearn.metrics
```

lineer modelin predict metodu ile tahminler yaptırabiliriz.

```python
predict = lin_model.predict(x_test)
```

```python
print(predict)
```

```python
for i in range(len(x_test)):
  print(f"{i} : gercek deger  : {y_test[i]} -- tahmin : {predict[i]}")
```

mae mse r2 bulma

```python
mae = sklearn.metrics.mean_absolute_error(y_test,predict)
mse = sklearn.metrics.mean_squared_error(y_test,predict)
r2 = sklearn.metrics.r2_score(y_test,predict)
```

```python
print(f"MAE : {mae}")
print(f"MSE : {mse}")
print(f"r2 : {r2}")
```
